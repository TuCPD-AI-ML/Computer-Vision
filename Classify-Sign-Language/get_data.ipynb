```python
import keras
import cv2
import numpy as np
import mediapipe as mp
```

```python
alp = ['A','B','C','D','E','F','G','H','I','K','L','M','N','O','P','Q','R','S','T','U','V','W','X','Y']
```

```python
model = keras.models.load_model('sign_lang3.h5')
```

```python
def classify(image):
    image = cv2.resize(image, (28, 28))
    image = image.astype("float") / 255.0
    image = np.reshape(image, (1, 28, 28, 1))
    pred=model.predict(image)
    idx = np.argmax(pred)
    return alp[idx]
```

```python
import time

cap = cv2.VideoCapture(0)
output_dir = "H:/sign_language/a"
output_hand = "H:/sign_language/hand"
images_per_second = 5
time_interval = 1 / images_per_second
save_img = False
image_count = 0
while 1:
    ret, img = cap.read()
    cv2.imshow("image", cv2.imread('signimage.png'))
    img = cv2.flip(img, 1)
    top, right, bottom, left = 85, 365, 325, 605
    roi = img[top:bottom, right:left]
    roi=cv2.flip(roi,1)
    gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (7, 7), 0)
    cv2.imshow('roi', blur)
    alpha=classify(blur)
    img_copy = img.copy()
    cv2.rectangle(img_copy, (left, top), (right, bottom), (0,255,0), 2)
    font=cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(img_copy,alpha,(0,130),font,5,(0,0,255),2)
    key = cv2.waitKey(1) & 0xFF
    if key == ord('o'):
        save_img = not save_img
        cv2.putText(img_copy,'start',(0,300),font,2,(0,0,255),2)
    if save_img:
        cv2.imwrite(f"{output_dir}/imageA_{time.time()}.png", roi)
        cv2.imwrite(f"{output_hand}/hand_{time.time()}.png", img)
        image_count += 1
        if image_count >= 50:
            break
        time.sleep(time_interval)
    cv2.imshow('img',img_copy)
        
    if key==ord('q'):
        break;
cap.release()
cv2.destroyAllWindows()
```
