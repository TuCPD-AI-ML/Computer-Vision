```python
import keras
import cv2
import numpy as np
```

```python
alp = ['A','B','C','D','E','F','G','H','I','K','L','M','N','O','P','Q','R','S','T','U','V','W','X','Y']
model = keras.models.load_model('sign_lang4_1.h5')
```

```python
def classify(image):
    image = cv2.resize(image, (28, 28))
    image = image.astype("float") / 255.0
    image = np.reshape(image, (1, 28, 28, 1))
    pred=model.predict(image)
    idx = np.argmax(pred)
    return alp[idx]
```

```python
cap = cv2.VideoCapture(0)
while 1:
    ret, img = cap.read()
    cv2.imshow("image", cv2.imread('signimage.png'))
    img = cv2.flip(img, 1)
    top, right, bottom, left = 85, 365, 325, 605
    roi = img[top:bottom, right:left]
    roi=cv2.flip(roi,1)
    gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (7, 7), 0)
    cv2.imshow('roi', blur)
    alpha = classify(blur)
    img_copy = img.copy()
    cv2.rectangle(img_copy, (left, top), (right, bottom), (0,255,0), 2)
    font=cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(img_copy,alpha,(0,130),font,5,(0,0,255),2)
    key = cv2.waitKey(1) & 0xFF
    cv2.imshow('img',img_copy)
        
    if key==ord('q'):
        break;
cap.release()
cv2.destroyAllWindows()
```
